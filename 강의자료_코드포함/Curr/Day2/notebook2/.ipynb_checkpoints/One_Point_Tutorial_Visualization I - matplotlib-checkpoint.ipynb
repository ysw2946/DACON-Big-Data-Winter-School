### 1.1. 가장 간단한 그래프 , <font color="royalblue">Line Plot</font>

가장 간단한 플롯은 선을 그리는 라인 플롯(line plot)이다.   
라인 플롯은 데이터가 시간, 순서 등에 따라 어떻게 변화하는지 보여주기 위해 사용한다.

>**사용 방법 :**  
>plt.`plot`(`x데이터`, `y데이터`)

#### 1.1.1. <font color="royalblue">첫번째 그래프</font> : 주어진 데이터가 하나일 때 (plot 함수의 데이터 인자가 1개일 때)

이런 경우, x축은 0, 1, 2, 3, ... 으로 자동 지정되고 넣은 데이터는 각 x축에 대응하는 y값이 된다.  
확인해보자!

```python
x = np.arange(1, 10)

plt.plot(x)
plt.show() #그래프를 보여줘!
```

(0, 1), (1, 2), ... , (8, 9)를 지나는 것을 확인할 수 있다.

#### 1.1.2. <font color="royalblue">두번째 그래프</font> : x, y축 데이터들이 모두 주어졌을 때

```python
x = np.arange(1, 10)
y = x*5

plt.plot(x, y)
plt.show() #그래프를 보여줘!
```

x,y 데이터가 (x, y) 좌표쌍으로 표현되어 그래프에 나타나는 것을 확인할 수 있다.

#### 1.1.3. <font color="royalblue">세번째 그래프</font> : 라인의 컬러 바꾸기

```python
x = np.arange(1, 10)
y = x*5

plt.plot(x, y, 'r') #위의 코드에서 'r'추가
plt.show() #그래프를 보여줘!
```

코드에 'r'을 추가한 것만으로 그래프가 빨간색으로 그려진 것을 확인할 수 있다.  
'r'은 red의 약자이고, 색을 지정하려면 plot에서 인자를 하나 추가해주면 된다.

### <div style=text-align:center>색상표</div>

|'r'|red|

|'w'|white|

#### 1.1.4. <font color="royalblue">네번째 그래프</font> : 라인 마커 변경하기

다양한 선의 종류를 선택할 수 있다. default 값은 직선이고, 점으로 표현하는 마커나 점선 등을 원한다면 선택할 수 있다.  
선의 종류는 plot에서 인자로 지정하면 되고, 색을 같이 지정하려면 같은 인자에 함께 문자형으로 색을 지정하면 된다.  
예시를 살펴보자! 동그란 마커 ‘o’를 붉은색 ‘r’로 표현하기 때문에, 세번째 인자를 ‘or’로 전달하였다.

```python
x = np.arange(1,10)
y = x*5

plt.plot(x,y,'or') #여기서 r은 red인데 o가 또 추가되었네!
plt.show()
```

위의 그래프들은 line plot이었는데 이번에는 선이 아닌 점으로 그려진 것을 확인할 수 있다.

#### 1.1.5. <font color="royalblue">다섯번째 그래프</font> : 축의 이름 / 그래프 제목 달기

>**사용 방법:**
>1. x축 이름 : plt.xlabel(`x축 이름`)
>2. y축 이름 : plt.ylabel(`y축 이름`)
>3. 그래프 제목 : plt.title(`그래프 제목`)

```python
x = np.arange(1, 10)
y = x*5

plt.plot(x, y)
plt.xlabel('x axis') # x축 이름 지정
plt.ylabel('y axis') # y축 이름 지정
plt.title('matplotlib sample') # 그래프 이름 지정

plt.show()
```

x축 이름/ y축 이름/ 그래프 제목이 생긴 것을 확인할 수 있다 !

#### 1.1.6. <font color="royalblue">여섯번째 그래프</font>: 구간 설정

x축과 y축의 구간을 설정할 수 있다.  

>**사용 방법:** <br>
>1. x축 구간 설정 : plt.xlim(시작 범위, 끝 범위)  
>2. y축 구간 설정 : plt.ylim(시작 범위, 끝 범위)

```python
plt.plot(x, y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('matplotlib sample')
plt.xlim(2, 3)
plt.ylim(5, 20)
plt.show()
```

x축은 2에서 3의 범위만, y축은 5에서 20의 범위만 나타내고 있는 것을 확인할 수 있다.

#### 1.1.7. <font color="royalblue">일곱번째 그래프</font>:  범주 달기 (legend)

여러 개의 그래프를 같이 그리는 경우가 존재하고, 이 경우 각 그래프를 구분하기 위해 그래프마다 라벨을 부여해 출력  
이를 범주(legend)라고 한다.

>**사용 방법:** <br>
>plt.legend(loc = `원하는 위치`) #위치 지정해주어야 함

```python
x = np.arange(1,10,0.1)
y1 = x*0.2
y2 = np.sin(x)

plt.plot(x,y1,'b',label='first') #파란색 / first 라벨 부여
plt.plot(x,y2,'r',label='second') #빨간색 / second 라벨 부여
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('matplotlib sample')
plt.legend(loc='upper right') # 범주 표현

plt.show()
```

#### 1.1.8. <font color="royalblue">여덟번째 그래프</font>:  어노테이션 (화살표로 가리키기)

그래프에 화살표를 그린후, 그 화살표에 문자열을 출력하는 기능

>**사용 방법:**  
>plt.annotate(`원하는 문자열`, xy(화살표 꼭지 xy좌표), xytext(문자열 xy 좌표), arrowprops(화살표 속성 ex.color))

```python
x = np.arange(1, 10)
y = x*5

plt.plot(x, y, 'b', label = 'total progress')
plt.xlabel('time')
plt.ylabel('achievement')
plt.title('Data Visualization AtoZ')
plt.legend(loc = 'lower right')
plt.annotate('here is our progress', xy = (1.5, 7.5), xytext = (3, 11), arrowprops = {'color': 'green'})
plt.show()
```

(1.5, 7.5)의 위치를 가리키는 화살표가 생긴 것을 확인할 수 있다.

#### 1.1.9. <font color="royalblue">아홉번째 그래프</font>: 서브 플롯 그리기
여러 개의 그래프를 그리고 싶을때 서브플롯을 사용

>**사용 방법:**  
>plt.subplot(`nrow`,`ncol`,`pos`)

```python
x = np.arange(1,10)
y1 = x*5
y2 = x*1
y3 = x*0.3
y4 = x*0.2

plt.subplot(2,2,1) #2행 2열 (2X2) 그래프 중 (1,1)의 위치
plt.plot(x,y1)

plt.subplot(2,2,2) #2행 2열 (2X2) 그래프 중 (1,2)의 위치
plt.plot(x,y2)

plt.subplot(2,2,3) #2행 2열 (2X2) 그래프 중 (2,1)의 위치
plt.plot(x,y3)

plt.subplot(2,2,4) #2행 2열 (2X2) 그래프 중 (2,2)의 위치
plt.plot(x,y4)

plt.show()
```

```python
그래프가 그려지는 순서는 행방향으로 진행된다는 것을 알 수 있다.
```

#### 1.1.10. <font color="royalblue">열번째 그래프</font>:  그래프 사이즈
>**사용 방법:**  
>plt.figure(figsize=(`가로`,`세로`))

```python
x = np.arange(1,10)
y1 = x*5
y2 = x*1
y3 = x*0.3
y4 = x*0.2

plt.figure(figsize=(20,5)) #가로 20, 세로 5로 설정

plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)

plt.show()
```

figsize를 어떤 숫자로 지정하느냐에 따라 그래프의 크기가 달라진다.

### 1.2. <font color="royalblue">Bar Plot</font>

라인 플롯은 ***연속적인 x***에 따른 y의 변화를 살펴보는 데에 효과적이라면, 막대 차트는 ***연속적이지 않은 x***에 따른 y의 변화를 살펴보는 데에 효과적이다.

>**사용 방법:**  
>1. plt.bar(`x`, `height`, `width`, `align`)  
**width:** 막대 두께 조절<br>
**align:** 'center'와 'edge' (기본값 'center')  
x축 label 막대의 중간 위치(center) 또는 왼쪽 가장자리 위치(edge) 설정
>2. `데이터`.plot(kind = 'bar')

#### 1.2.1. <font color="royalblue">첫번째 그래프</font>:  첫번째 방법으로 막대 차트 그리기

```python
x = np.random.randint(1, 6, size = 100)
uni, count = np.unique(x, return_counts = True) # 각 빈도수를 count에 반환

plt.bar(uni, count, color = 'red')
plt.show()
```

```python
plt.bar(uni, count, width = 0.3, color="green") #width 조절
plt.show()
```

```python
plt.bar(uni, count, align='edge', color="k") #align 설정
plt.show()
```

#### 1.2.1. <font color="royalblue">두번째 그래프</font>: 막대 방향을 수평으로 바꾸기

```python
plt.barh(uni, count, color="royalblue") # bar에서 barh로!
plt.show()
```

#### 1.2.1. <font color="royalblue">세번째 그래프</font>: 양쪽으로 막대차트 그리기

```python
x=np.arange(12).reshape((4, 3))
plt.barh(x[:,0], x[:,1], color='g', height=2 )
plt.barh(x[:,0], -x[:, 2], color='royalblue', height=2)
plt.show()
```

#### 1.2.1. <font color="royalblue">네번째 그래프</font>: 두번째 방법으로 막대 차트 그리기

```python
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))

data.plot(kind='bar')
plt.show()
```

#### 1.2.1. <font color="royalblue">다섯번째 그래프</font>: 두번째 방법으로 막대 차트 그리기

```python
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))

data.plot(kind='bar')
plt.show()
```

#### 1.2.1. <font color="royalblue">여섯번째 그래프</font>: 막대 방향 수평으로 바꾸기

```python
data.plot(kind='barh') # bar에서 barh로 !
plt.show()
```

#### 1.2.1. <font color="royalblue">일곱번째 그래프</font>: data.frame을 그래프로 그리기

```python
df2 = pd.DataFrame(np.random.rand(6,4),
                  index = ['one','two','three','four','five','six'],
                  columns = pd.Index(['A','B','C','D'], name='column'))
df2.head()
```

```python
df2.plot(kind='bar')
plt.show()
```

#### 1.2.1. <font color="royalblue">여덟번째 그래프</font>: Stacked Bar Chart

```python
df2.plot(kind='bar', stacked=True)
plt.show()
```

```python
df2.plot(kind='barh', stacked=True) # 막대 수평방향으로 쌓는 형식으로 보여줘!
plt.show()
```

### 1.3. <font color="royalblue">Histogram</font>

히스토그램은 확률분포의 그래픽적인 표현으로 막대그래프 중 한 종류.
확률분포와 관계가 있으므로 통계적 요소를 나타내기 위해 많이 사용한다.

>**사용 방법:**  
>plt.hist(`data`, bins = `숫자`, density = True or False(False가 default))

`bins`는 구간의 개수를 의미하고, `density=True`는 빈도수가 아닌 비율로 나타내는 것을 의미한다.

```python
x=np.random.randn(1000)
plt.hist(x, color='skyblue') #bins default는 10
plt.show()
```

```python
x=np.random.randn(1000)
plt.hist(x, color='grey', bins=50) #구간의 개수를 50개
plt.show()
```

```python
plt.hist(x, density=True, color='lightpink') #빈도수가 아닌 확률로!
plt.show()
```

1차원의 데이터를 인자로 전달하면 데이터의 상대적 면적을 자동으로 계산하여 그래프를 그려주는 것이 파이차트이다.<br>
단일 범주형 변수 대상으로 구성비를 확인하는 데 유용하다.

>**사용 방법:**  
>plt.pie(`data`)

```python
plt.pie(np.arange(101))
plt.show()
```
